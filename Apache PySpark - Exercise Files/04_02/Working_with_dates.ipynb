```python
from pyspark.sql.functions import to_timestamp,col,lit
rc = spark.read.csv('reported-crimes.csv',header=True).withColumn('Date',to_timestamp(col('Date'),'MM/dd/yyyy hh:mm:ss a')).filter(col('Date') < lit('2018-11-12'))
rc.show(5)
```

## Working with dates

```python
from pyspark.sql.functions import to_date, to_timestamp, lit
```

```python
rc.select(to_date(lit('2019-12-25')), to_date(lit('2019-25-15'))).show(1)
```

  **2019-12-25 13:30:00**

```python
df = spark.createDataFrame([('2019-12-25 13:30:00',)], ['Christmas'])
df.show(1)
```

```python
df.select(to_date('Christmas', 'yyyy-MM-dd HH:mm:ss'), to_timestamp('Christmas', 'yyyy-MM-dd HH:mm:ss')).show()
```

**25/Dec/2019 13:30:00**

```python
df = spark.createDataFrame([('25/Dec/2019 13:30:00',)], ['Christmas'])
df.show(1)
```

```python
df.select(to_date('Christmas', 'dd/MMM/yyyy HH:mm:ss'), to_timestamp('Christmas', 'dd/MMM/yyyy HH:mm:ss')).show()
```

**12/25/2019 01:30:00 PM**

```python
df = spark.createDataFrame([('12/25/2019 01:30:00 PM',)], ['Christmas'])
df.show(1)
```

```python
df.select(to_date('Christmas', 'MM/dd/yyyy hh:mm:ss aa'), to_timestamp('Christmas', 'MM/dd/yyyy hh:mm:ss aa')).show()
```
